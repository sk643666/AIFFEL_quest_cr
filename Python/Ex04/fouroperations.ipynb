```python
import sys  #    sys  .

#    
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  if b == 0:
    raise ZeroDivisionError(' 0   .')   # 0     ->   ZeroDivisionError 
  return f' : {a // b},  : {a % b},  : {a / b}'

#      
while True:
  try:

    #     
    a = int(input('    : '))
    b = int(input('    : '))

    operation = input('  (+, -, *, /) : ')

    #     
    if operation == '+':
      print(f'  : {add(a,b)}')
    elif operation == '-':
      print(f'  : {subtract(a,b)}')
    elif operation == '*':
      print(f'  : {multiply(a,b)}')
    elif operation == '/':
      print(f'  : {divide(a,b)}')
    else:
      print('  .  .')

  #      
  except ValueError:
    print('Error :    .  .')

  #  0   
  except ZeroDivisionError as e:
    print(f'Error : {e}  .')

  #        .
  while True:
    continue_cal = input('   ? (y/n) : ')
    if continue_cal == 'y':
      break
      # 'y'     .

    elif continue_cal == 'n':
      print(' .')
      sys.exit()
      # 'n'   .

    else:
      print(' .  .')
```
