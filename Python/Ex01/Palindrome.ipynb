위에서 이스케이프문자 관련한 것을 좀 고민해 보았습니다... (쉽지않네요)

\t가 string 안에 포함된경우 if 문 들어가야 할 것 같은데...
**\\t로 해야 되더군요.. 왜일까요?**

raw_string 의 escape+t와 지금 \\t 가 같은 것이다?

```python
def function(s):
  if "\\t" in s:

    s = s.replace("\\t", "")

  return s

a = input("회문 입력")

tmp = function(a)

if tmp == tmp[::-1]:
  print('회문')

else:
  print('not 회문')
```
